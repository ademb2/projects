```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('teeth_characteristics.csv')
df.head()
```

```python

# Split the values in the "dental formula" column and assign them to new columns
df[['IM', 'CM', 'PM', 'MM', 'Im', 'Cm', 'pM', 'Mm']] = df['Dental formula'].str.split('[./]', expand=True)

df
```

```python
import requests
from bs4 import BeautifulSoup

def scrape_taxonomy(species):
    url = f"https://animaldiversity.org/accounts/{species}/classification/#{species}"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    taxonomy = []
    taxon_rows = soup.find_all('div', class_='taxon-row')
    for row in taxon_rows:
        rank = row.find('span', class_='rank').text
        name = row.find('a', class_='taxon-link').text
        vernacular_name = row.find('span', class_='vernacular-name').text
        taxonomy.append((rank, name, vernacular_name))
    
    return taxonomy

species_list = ["Ovis_canadensis", "Canis_lupus", "Felis_catus"]  # Replace with your list of species

for species in species_list:
    taxonomy = scrape_taxonomy(species)
    print(f"Taxonomy for {species}:")
    for rank, name, vernacular_name in taxonomy:
        print(f"{rank}: {name} ({vernacular_name})")
    print()
```
