```python
species_names=("Loxodonta africana", "Ovis aries", "Bos taurus", "Alces alces", "Antilocapra americana", "Bison bison", "Odocoileus virginianus", "Oreamnos americanus", "Ovibos moschatus", "Ovis canadensis", "Dasypus novemcinctus", "Cervus canadensis", "Rangifer tarandus", "Antrozous pallidus", "Lasiurus blossevillii", "Leptonycteris curasoae", "Propithecus verreauxi", "Callithrix jacchus", "Daubentonia madagascariensis", "Sorex arcticus", "Enhydra lutris", "Lama glama", "Lama guanicoe", "Vicugna vicugna", "Vicugna pacos", "Homo sapiens", "Tarsius syrichta", "Mephitis mephitis", "Mustela spp.", "Taxidea taxus", "Neovison vison", "Lemur catta", "Galago senegalensis", "Nycticebus coucang", "Sapajus spp.", "Tayassuidae spp.", "Martes spp.", "Procyon lotor", "Nasua nasua", "Equus caballus", "Sus scrofa", "Sus scrofa", "Didelphis virginiana", "Mus musculus", "Ondatra zibethicus", "Oryzomys palustris", "Zapus hudsonius", "Castor canadensis", "Dipodomys ingens", "Erethizon dorsatum", "Geomys spp.", "Myocastor coypus", "Sciurus apache", "Aplodontia rufa", "Sciurus granatensis", "Tamias spp.", "Tamiasciurus hudsonicus", "Cynomys ludovicianus", "Marmota monax", "Ochotona princeps", "Oryctolagus cuniculus", "Lepus europaeus", "Lynx canadensis", "Panthera onca", "Felis catus", "Panthera leo", "Leopardus pardalis", "Vulpes lagopus", "Urocyon cinereoargenteus", "Canis latrans", "Ursus arctos", "Ursus maritimus", "Ursus arctos horribilis", "Ursus americanus", "Canis lupus", "Canis lupus familiaris")
```

```python
df = df.assign(Species_Name=species_names)
df
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Initialize a list to store data dictionaries for each species
data_list = []

# Assuming 'species_names' is a list of species names
for species_name in species_names:
    # Format the URL with the species name
    url = f"https://animaldiversity.org/accounts/{species_name}"

    # Retrieve the HTML content
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    # Find all h3 elements with id="habitat"
    habitat_headers = soup.find_all("h3", id="habitat")

    for habitat_header in habitat_headers:
        # Extract habitat text
        habitat_text = habitat_header.text.strip()

        # Find all ul elements directly following the habitat header
        ul_elements = habitat_header.find_next_siblings("ul")

        # Loop through each ul element (assuming multiple might exist)
        for ul_element in ul_elements:
            # Lists to store keywords and gloss data
            keywords = []
            gloss_texts = []

            # Loop through keywords-header elements within the current ul
            for keyword_element in ul_element.find_all("li", class_="keywords-header"):
                keywords.append(keyword_element.text.strip())

            # Loop through gloss elements within the current ul
            for gloss_link in ul_element.find_all("a", class_="gloss"):
                gloss_texts.append(gloss_link.text.strip())

            # Loop through span elements within the current ul
            for span_element in ul_element.find_all("span"):
                gloss_texts.append(span_element.text.strip())

            # Convert gloss_texts to a string representation
            gloss_texts_str = ', '.join(gloss_texts)

            # Create a dictionary for the current species and habitat
            species_data = {'Species Name': species_name, 'Habitat': habitat_text}

            # Add keywords and gloss texts to the dictionary
            for k in keywords:
                species_data[k] = gloss_texts_str

            # Append the data dictionary for the current species to the list
            data_list.append(species_data)

    # Find all h3 elements with id="food_habits"
    food_habit_headers = soup.find_all("h3", id="food_habits")

    for food_habit_header in food_habit_headers:
        # Extract food habit text
        food_habit_text = food_habit_header.text.strip()

        # Find all ul elements directly following the food habit header
        ul_elements = food_habit_header.find_next_siblings("ul")

        # Loop through each ul element (assuming multiple might exist)
        for ul_element in ul_elements:
            # Lists to store keywords and gloss data
            keywords = []
            gloss_texts = []

            # Loop through keywords-header elements within the current ul
            for keyword_element in ul_element.find_all("li", class_="keywords-header"):
                keywords.append(keyword_element.text.strip())

            # Loop through gloss elements within the current ul
            for gloss_link in ul_element.find_all("a", class_="gloss"):
                gloss_texts.append(gloss_link.text.strip())

            # Loop through span elements within the current ul
            for span_element in ul_element.find_all("span"):
                gloss_texts.append(span_element.text.strip())

            # Convert gloss_texts to a string representation
            gloss_texts_str = ', '.join(gloss_texts)

            # Update the data dictionary for the current species with food habits
            species_data['Food Habits'] = food_habit_text

            # Add keywords and gloss texts related to food habits to the dictionary
            for k in keywords:
                species_data[k] = gloss_texts_str

            # Append the updated data dictionary for the current species to the list
            data_list.append(species_data)

# Create a DataFrame from the collected data
df = pd.DataFrame(data_list)
```
