```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Initialize empty lists for column names and values
column_names = []
column_values = []

# Iterate over each species name
for species_name in species_names:
    # Format the URL with the species name
    url = f"https://animaldiversity.org/accounts/{species_name}"
    
    # Send a GET request to the URL
    response = requests.get(url)
    
    # Create a BeautifulSoup object from the response content
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find all the ul elements with class "keywords"
    ul_elements = soup.find_all('ul', class_='keywords')
    
    # Iterate over each ul element
    for ul in ul_elements:
        # Find the li element with class "keywords-header" to get the column name
        column_name = ul.find('li', class_='keywords-header').text.strip()
        column_names.append(column_name)
        
        # Find all the span elements within the ul element to get the column values
        span_elements = ul.find_all('span')
        values = [span.text.strip() for span in span_elements]
        column_values.append(', '.join(values))

# Create a dictionary from the column names and values
data = dict(zip(column_names, column_values))

# Create a DataFrame from the dictionary with an index
df = pd.DataFrame(data, index=[0])

```
